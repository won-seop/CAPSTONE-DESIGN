```python
import random

for index, row in filtered_df.iterrows():
  nouns = []
  for noun in hannanum.nouns(row['title_data']):
    if noun not in stop_words:
      nouns.append(noun)
  if len(nouns) < 4:
    filtered_df.at[index, 'keyword'] = ', '.join(nouns)
  elif len(nouns) >= 4:
    nouns = random.sample(nouns, 4)
    filtered_df.at[index, 'keyword'] = ', '.join(nouns)

filtered_df
```
